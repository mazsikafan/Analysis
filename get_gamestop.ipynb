```python
import os
import pandas as pd

```

```python
def extract_data_to_dataframe(directory):
    data = {
        'SETTLEMENT DATE': [],
        'QUANTITY (FAILS)': [],
        'PRICE': []
    }

    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            filepath = os.path.join(directory, filename)
            with open(filepath, 'r', encoding='ISO-8859-1') as file:  # change encoding here
                for line in file:
                    fields = line.strip().split('|')
                    if len(fields) >= 6:
                        data['SETTLEMENT DATE'].append(fields[0])
                        data['QUANTITY (FAILS)'].append(fields[3])
                        data['PRICE'].append(fields[5])

    df = pd.DataFrame(data)
    return df

directory_path = '/Users/marotibalint/Documents/options_pricing/ftd'
df = extract_data_to_dataframe(directory_path)
```
